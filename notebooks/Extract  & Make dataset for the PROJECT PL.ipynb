```python
import soccerdata as sd
import pandas as pd
```

```python
seasons = [2021, 2022, 2023, 2024, 2025]
all_games_dfs = []
for x in seasons:
  mh = sd.MatchHistory(leagues="ENG-Premier League", seasons=x)
  games_df = mh.read_games()
  all_games_dfs.append(games_df)

df = pd.concat(all_games_dfs)
```

```python
df.tail()
```

```python
columns_to_keep = [
    "date", "home_team", "away_team", "FTHG", "FTAG",
    "HS", "AS", "HST", "AST",
    "HF", "AF", "HC", "AC",
    "HY", "AY", "HR", "AR",
    "HTHG", "HTAG"
]

df_pl = df[columns_to_keep].copy()

df_pl = df_pl.rename(columns={
    "FTHG": "home_goals",
    "FTAG": "away_goals",
    "HS": "home_shots",
    "AS": "away_shots",
    "HST": "home_shots_on_target",
    "AST": "away_shots_on_target",
    "HF": "home_fouls",
    "AF": "away_fouls",
    "HC": "home_corners",
    "AC": "away_corners",
    "HY": "home_yellow",
    "AY": "away_yellow",
    "HR": "home_red",
    "AR": "away_red",
    "HTHG": "home_ht_goals",
    "HTAG": "away_ht_goals"
})
```

```python
df_pl.head()
```

```python
df_pl.info()
```

```python
df_pl['time'] = df_pl['date'].dt.time
```

```python
df_pl['date'] = df_pl['date'].dt.date
```

```python
df_pl.head()
```

```python
df_pl = df_pl.sort_values(by=['date', 'time'])
```

```python
df_pl.reset_index(drop = True, inplace= True)
```

```python
df_pl.head()
```

```python
df_pl.to_csv('Premier_league_data_2021_to_2025.csv')
```
